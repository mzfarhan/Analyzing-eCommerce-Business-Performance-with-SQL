```python
import pandas as pd
import numpy as np
```

**geolocation_dataset Preprocessing**

```python
dfg = pd.read_csv('geolocation_dataset.csv')
dfg.head()
```

```python
# Delete duplicate geolocation_zip_code_prefix
dfg.drop_duplicates('geolocation_zip_code_prefix', keep='first', inplace=True)
dfg.to_csv('filtered_geolocation_dataset.csv', index=False)
dfg.head()
```

**customers_dataset Preprocessing**

```python
dfc = pd.read_csv('customers_dataset.csv')
dfc.head()
```

```python
# Filtering data customer_zip_code_prefix based on the data available in geolocation_zip_code_prefix
filtered_dfc = dfc[dfc['customer_zip_code_prefix'].isin(dfg['geolocation_zip_code_prefix'])]
filtered_dfc.to_csv('filtered_customers_dataset.csv', index=False)
filtered_dfc.head()
```

**sellers_dataset Proprocessing**

```python
dfs = pd.read_csv('sellers_dataset.csv')
dfs.head()
```

```python
# Filtering data seller_zip_code_prefix based on the data available in geolocation_zip_code_prefix
filtered_dfs = dfs[dfs['seller_zip_code_prefix'].isin(dfg['geolocation_zip_code_prefix'])]
filtered_dfs.to_csv('filtered_sellers_dataset.csv', index=False)
filtered_dfs.head()
```

**orders_dataset Preprocessing**

```python
dfo = pd.read_csv('orders_dataset.csv')
dfo.head()
```

**order_payments_dataset Preprocessing**

```python
dfop = pd.read_csv('order_payments_dataset.csv')
dfop.head()
```

```python
# Filtering data order_id in order_payments_dataset based on the data available in order_id in orders_dataset
filtered_dfop = dfop[dfop['order_id'].isin(filtered_dfo['order_id'])]
filtered_dfop.to_csv('filtered_order_payments_dataset.csv', index=False)
filtered_dfop.head()
```

**order_reviews_dataset Preprocessing**

```python
dfor = pd.read_csv('order_reviews_dataset.csv')
dfor.head()
```

```python
# Filtering data order_id in order_reviews_dataset based on the data available in order_id in orders_dataset
filtered_dfor = dfor[dfor['order_id'].isin(filtered_dfo['order_id'])]
filtered_dfor.to_csv('filtered_order_reviews_dataset.csv', index=False)
filtered_dfor.head()
```
